4. Program to calculate the average of 5 numbers

```python
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
num3 = float(input("Enter third number: "))
num4 = float(input("Enter fourth number: "))
num5 = float(input("Enter fifth number: "))
average = (num1 + num2 + num3 + num4 + num5) / 5
print(f"The average of the five numbers is {average}.")
```

5. Program to calculate percentage from given marks

```python
total_marks = float(input("Enter the total marks: "))
obtained_marks = float(input("Enter the obtained marks: "))
percentage = (obtained_marks / total_marks) * 100
print(f"The percentage is {percentage:.2f}%.")
```

7. Program to calculate the area of a circle

10. Program to convert temperature from Celsius to Fahrenheit

```python
celsius = float(input("Enter temperature in Celsius: "))
fahrenheit = (celsius * 9/5) + 32
print(f"{celsius}°C is equal to {fahrenheit}°F.")
```

11. Program to implement formula (b² - 4ac) / 2a

```python
a = float(input("Enter the value of a: "))
b = float(input("Enter the value of b: "))
c = float(input("Enter the value of c: "))
result = (b ** 2 - 4 * a * c) / (2 * a)
print(f"The result of the formula (b² - 4ac) / 2a is {result}.")
```

12. Scenario: Calculating the Total Cost with Discount for 13 pens

```python
price_per_pen = 75
num_pens = 13
total_cost = price_per_pen * num_pens
discount = 0.05 * total_cost
final_cost = total_cost - discount
print(f"The total amount to pay after a 5% discount is {final_cost:.2f} Rs.")
```

13. Scenario: Calculating the Total Cost with Bulk Purchase Discount for Books

```python
price_per_book = 150
num_books = int(input("Enter the number of books: "))
total_cost = price_per_book * num_books
if num_books >= 10:
    discount = 0.10 * total_cost  # 10% discount
    total_cost -= discount
print(f"The total amount to pay is {total_cost:.2f} Rs.")
```

14. Program to swap two numbers without using a temporary variable

```python
x = float(input("Enter the first number (x): "))
y = float(input("Enter the second number (y): "))
x, y = y, x
print(f"After swapping, x = {x}, y = {y}.")
```

15. Program to divide books equally among students and find the remainder

```python
K = int(input("Enter the total number of books (K): "))
N = int(input("Enter the total number of students (N): "))
books_per_student = K // N
remaining_books = K % N
print(f"Each student gets {books_per_student} books.")
print(f"The remaining number of books is {remaining_books}.")
```
