5. Program to Calculate Shopping Discount Based on Total Bill Amount in INR:
Write a program that calculates the discount on a shopping bill based on the total amount
spent. The discount rules are as follows:
a. If the total bill is less than ₹1,000, no discount is applied.
b. If the total bill is between ₹1,000 and ₹5,000, a 10% discount is applied.
c. If the total bill is between ₹5,000 and ₹10,000, a 15% discount is applied.
d. If the total bill is above ₹10,000, a 20% discount is applied.

```python
total_bill = float(input("Enter the total bill amount in INR: "))

if total_bill < 1000:
    discount = 0
elif total_bill <= 5000:
    discount = 0.10 * total_bill
elif total_bill <= 10000:
    discount = 0.15 * total_bill
else:
    discount = 0.20 * total_bill

final_bill = total_bill - discount
print(f"The final amount after discount is ₹{final_bill:.2f}.")
```

8. Program to calculate income tax based on age and gender:
Write a program that calculates income tax based on the following rules:
o Males under 60 have a tax rate of 10% on income above a certain threshold.
o Females under 60 have a tax rate of 8% on income above a certain threshold.
o Senior citizens (both male and female) have a tax exemption up to a higher
threshold, and any income above that is taxed at 5%.

```python
age = int(input("Enter your age: "))
gender = input("Enter your gender (male/female): ").lower()
income = float(input("Enter your annual income in INR: "))

threshold = 250000 if age < 60 else 500000

if age < 60:
    if gender == "male":
        tax_rate = 0.10
    else:
        tax_rate = 0.08
else:
    tax_rate = 0.05

taxable_income = max(0, income - threshold)
tax = tax_rate * taxable_income
print(f"Your income tax is ₹{tax:.2f}.")
```

9. Program to determine the ticket price for public transport:
A public transport system offers different ticket prices based on age and gender:
o Children under 12 get a 50% discount.
o Senior citizens above 60 get a 30% discount.
o Women get a 10% discount.
o Write a program that takes age, gender, and the base ticket price as input and
calculates the final ticket price after applying all relevant discounts.

```python
age = int(input("Enter your age: "))
gender = input("Enter your gender (male/female): ").lower()
base_price = float(input("Enter the base ticket price: "))

discount = 0
if age < 12:
    discount = 0.50
elif age > 60:
    discount = 0.30
elif gender == "female":
    discount = 0.10

final_price = base_price * (1 - discount)
print(f"The final ticket price is ₹{final_price:.2f}.")
```

11. Program to determine eligibility for a student scholarship:
The government provides scholarships to students based on their age, gender, and
family income:
o Students under 18 are eligible for a junior scholarship.
o Female students receive a bonus amount on top of the scholarship.
o Family income below a certain threshold qualifies the student for additional
financial aid.
Write a program that takes the student’s age, gender, and family income as input and
determines their eligibility and scholarship amount.

```python
age = int(input("Enter the student's age: "))
gender = input("Enter the student's gender (male/female): ").lower()
family_income = float(input("Enter the student's family income in INR: "))

if age < 18:
    scholarship_amount = 10000
    if family_income < 500000:
        scholarship_amount += 5000  # Additional financial aid
    if gender == "female":
        scholarship_amount += 2000  # Bonus for female students
    print(f"The student is eligible for a scholarship of ₹{scholarship_amount}.")
else:
    print("The student is not eligible for a junior scholarship.")
```

14. Write a program to calculate the electricity bill (accept number of unit from user)
according to the following criteria :
Unit Price
First 100 units no charge
Next 100 units Rs 5 per unit
After 200 units Rs 10 per unit
Example 1:
input: unit is 350
Output: total bill amount is Rs2000
Example 2:
input: unit is 250
Output: total bill amount is Rs1000

Example 3:
input: unit is 190
Output: total bill amount is Rs450

```python
units = int(input("Enter the number of units used: "))

if units <= 100:
    bill = 0
elif units <= 200:
    bill = (units - 100) * 5
else:
    bill = (100 * 5) + (units - 200) * 10

print(f"The total bill amount is ₹{bill}.")
```

according to the following criteria
Upto 2.5 lakh- No tax
Above 2.5lakh to 5lakh: 5percent tax
Above 5 lakh to 10lakh-20 percent
Above 10lakh:30 percent
Example 1:
input: 400000
Output: 7500
Explaination: upto 2.5lakh there is no tax.
Remaining amount is 150000 so 5 percent of it 7500.
Example 2:

Output: 32500
Explaination:

Example 3:
input: unit is 900000
Output: 92500
Explaination: tax upto 5 lakh income is 12500
Above that amount is 4lakh.so by 20percent slab tax on 4 lakh is 80000.So total tax is
:92500

Example 4:
input: unit is 1200000
Output:172500

Tax upto 5 lakh: 12500

Remaining amount:200000 is in 30percent slab,so tax on it is 60000
